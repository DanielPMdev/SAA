# **Credit Card**

https://www.kaggle.com/arjunbhasin2013/ccdata?select=CC+GENERAL.csv


This case requires to develop a customer segmentation to define marketing strategy. The sample Dataset summarizes the usage behavior of about 9000 active credit card holders during the last 6 months. The file is at a customer level with 18 behavioral variables.

Following is the Data Dictionary for Credit Card dataset:

*   CUSTID: Identification of Credit Card holder (Categorical)
*   BALANCE: Balance amount left in their account to make purchases
*   BALANCEFREQUENCY: How frequently the Balance is updated, score between 0 and 1 (1 = frequently updated, 0 = not frequently updated)
*   PURCHASES: Amount of purchases made from account
*   ONEOFFPURCHASES: Maximum purchase amount done in one-go
*   INSTALLMENTSPURCHASES: Amount of purchase done in installment
*   CASHADVANCE: Cash in advance given by the user
*   PURCHASESFREQUENCY: How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased, 0 = not frequently purchased)
*   ONEOFFPURCHASESFREQUENCY: How frequently Purchases are happening in one-go (1 = frequently purchased, 0 = not frequently purchased)
*   PURCHASESINSTALLMENTSFREQUENCY: How frequently purchases in installments are being done (1 = frequently done, 0 = not frequently done)
*   CASHADVANCEFREQUENCY: How frequently the cash in advance being paid
*   CASHADVANCETRX: Number of Transactions made with "Cash in Advanced"
*   PURCHASESTRX: Numbe of purchase transactions made
*   CREDITLIMIT: Limit of Credit Card for user
*   PAYMENTS: Amount of Payment done by user
*   MINIMUM_PAYMENTS: Minimum amount of payments made by user
*   PRCFULLPAYMENT: Percent of full payment paid by user
*   TENURE: Tenure of credit card service for u

# **Limpieza de datos**

## **Valores nulos**

```python
dataset.isna().sum()
```

```python
dataset['MINIMUM_PAYMENTS'] = dataset['MINIMUM_PAYMENTS'].fillna(dataset['MINIMUM_PAYMENTS'].mean())
dataset['CREDIT_LIMIT'] = dataset['CREDIT_LIMIT'].fillna(dataset['CREDIT_LIMIT'].mean())
dataset.isna().sum()
```

## **Outliers (a realizar)**

```python
# Visualización rápida de outliers
plt.figure(figsize=(20,10))
sns.boxplot(data=dataset)
plt.xticks(rotation=90)
plt.title("Distribución de variables y Outliers")
plt.show()

for col in dataset.columns:
    if col != 'CUST_ID':
        Q1 = dataset[col].quantile(0.25)
        Q3 = dataset[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        dataset[col] = np.clip(dataset[col], lower_limit, upper_limit)
```

# **Reducción dimensionalidad (a realizar)**

```python
dataset.drop(['CUST_ID'], axis=1, inplace=True)
dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Filtramos para quedarnos SOLO con los números (esto ignora CUST_ID automáticamente)
dataset_numeric = dataset.select_dtypes(include=[np.number])

# 2. Escalado de datos usando solo los números
scaler = StandardScaler()
dataset_scaled = scaler.fit_transform(dataset_numeric)

# 3. Aplicación de PCA
pca = PCA()
pca.fit(dataset_scaled)

# Visualización de la varianza explicada acumulada
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Análisis de PCA')
plt.grid()
plt.show()
```

```python
pca_final = PCA(n_components=2)
dataset_pca = pca_final.fit_transform(dataset_scaled)
df_pca = pd.DataFrame(data=dataset_pca, columns=['PC1', 'PC2'])
```

## **Varianza**

```python
from sklearn.feature_selection import VarianceThreshold

# Calculamos la varianza de cada columna
variances = dataset.var()
print("Varianza de las variables:\n", variances.sort_values())
```

## **Matriz de Correlación**

```python
plt.figure(figsize=(15, 8))
sns.heatmap(round(dataset.corr(method='spearman'), 2), annot=True, mask=None)
plt.show()
```

## **PCA**

```python
# Crear un DataFrame con los componentes del PCA
comps = pd.DataFrame(pca_final.components_, columns=dataset_numeric.columns)

# Visualizar qué variables influyen más en cada componente
plt.figure(figsize=(20,5))
sns.heatmap(comps, cmap='coolwarm', annot=True)
plt.title('Influencia de las variables originales en PC1 y PC2')
plt.show()

# PC1 suele representar el "Gasto/Balance General".
# PC2 suele representar la "Frecuencia o Tipo de Compra (Contado vs Plazos)".
```

# **Clustering**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Determinando el número de clusters (Método del Codo)
wcss = [] # Within-Cluster Sum of Square
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(dataset_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='s')
plt.title('Método del Codo (Elbow Method)')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(2,15))
visualizer.fit(dataset_scaled)
visualizer.poof()
```

```python
# Aplicando K-Means con el número óptimo
k_optimo = 4
kmeans = KMeans(n_clusters=k_optimo, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(dataset_scaled)

# Añadir la etiqueta de cluster al dataset original
dataset['Cluster'] = clusters

# Calcular el Score de Silueta
score = silhouette_score(dataset_scaled, clusters)
print(f'Silhouette Score para k={k_optimo}: {score:.3f}')
```

```python
# Visualización de los clusters en los componentes de PCA
plt.figure(figsize=(10,7))
sns.scatterplot(x=df_pca['PC1'], y=df_pca['PC2'], hue=clusters, palette='viridis')
plt.title(f'Segmentación de Clientes (K={k_optimo}) en espacio PCA')
plt.show()
```

```python
# Resumen de los segmentos
print(dataset.groupby('Cluster').mean())
```

```python
# Visualización del perfil de cada Cluster
# Seleccionamos algunas columnas clave para analizar el comportamiento
best_cols = ['BALANCE', 'PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT', 'PAYMENTS', 'PRC_FULL_PAYMENT']

# Creamos un pairplot o boxplots para ver las diferencias
for col in best_cols:
    if col in dataset.columns:
        plt.figure(figsize=(10,5))
        sns.boxplot(x='Cluster', y=col, data=dataset)
        plt.title(f'Distribución de {col} por Cluster')
        plt.show()
```

# **Clustering Jerárquico (Método Alternativo a K-means)**

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

# 1. Visualización del Dendrograma
# Usamos una muestra de 200 datos para que el gráfico sea legible y rápido, sino el gráfico sería una mancha negra ininteligible.
plt.figure(figsize=(15, 8))
plt.title('Dendrograma (Muestra de 200 clientes)')
plt.xlabel('Clientes')
plt.ylabel('Distancia Euclídea')

# Calculamos el dendrograma sobre una muestra reducida
dendrogram = sch.dendrogram(sch.linkage(dataset_scaled[:200], method='ward'))
plt.show()
```

Explicación:


*   Las líneas verticales representan clientes o grupos.
*   La altura de las líneas horizontales representa la distancia (diferencia) entre grupos.
*   Una línea horizontal larga sin cortes sugiere una buena separación natural



```python
# 2. Entrenando el modelo Jerárquico
hc = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')

y_hc = hc.fit_predict(dataset_scaled)
```

```python
# 3. Visualización de los resultados usando PCA
dataset['Cluster_Jerarquico'] = y_hc

plt.figure(figsize=(10,7))
sns.scatterplot(x=df_pca['PC1'], y=df_pca['PC2'], hue=dataset['Cluster_Jerarquico'], palette='cool')
plt.title('Segmentación Jerárquica en espacio PCA')
plt.show()
```

```python
# Comparación rápida de tamaños de grupos
print("Distribución de clientes por grupo (Jerárquico):")
print(dataset['Cluster_Jerarquico'].value_counts())
```

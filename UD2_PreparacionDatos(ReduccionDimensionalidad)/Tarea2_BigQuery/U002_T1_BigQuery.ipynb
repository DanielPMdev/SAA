## Categorizar el signal

```python
def clasificar_signal(valor):
    if valor == 0:
        return 1
    elif 1 <= valor <= 5:
        return 2
    elif 6 <= valor <= 10:
        return 3
    elif 11 <= valor <= 15:
        return 4
    elif 16 <= valor <= 25:
        return 5
    else:
        return 0

df['signal'] = df['signal'].apply(clasificar_signal)
```
